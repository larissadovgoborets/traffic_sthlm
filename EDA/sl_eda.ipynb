```python
import requests
import pprint

# Define the API endpoint
# The full URL
# "https://deviations.integration.sl.se/v1/messages?future=<true/false>&site=<siteId1,siteId2>&line=<lineId1,lineId2>&transport_mode=<mode1,mode2>"

# The URL to include the future messages and not filter by other parameters
api_url = "https://deviations.integration.sl.se/v1/messages?future=true"

# Make a GET request to fetch the data
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Load the data into a pandas DataFrame
    data = response.json()
    print("Data fetched successfully!")
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")

# Display the first few rows of the DataFrame

pprint.pprint(data[:3])
```
